```python
def distance(ll_pair):
    """calculates haversine distance for list of (lat, lon) pairs
    :params ll_pair: numpy array of (lat, lon) pairs
    :returns: matrix with pairwise distances
    """
    pos = np.deg2rad(ll_pair)
    #lat_diff, lon_diff = np.diff(pos, axis=0).T
    lat, lon = ll_pair.T
    lat_diff = lat[:, None] - lat
    lon_diff = lon[:, None] - lon
    d = np.sin(lat_diff/2)**2 + np.cos(lat[:, None]) * np.cos(lat) * np.sin(lon_diff/2)**2
    return 2 * 6371 * np.arcsin(np.sqrt(d))
```

## Clustering

I shold npt use all 13k unique sensors for the distance function because it generates a 13k*13k matrix with a lot of combinations. Therefore, I cluster first and calculate distance for each cluster.

```python
df.loc[:, ['lat', 'lon']].values
```

```python
df.head()
```

```python
# find suitable number of clusters
res = []
n = np.arange(2, 40, 2)
for i in n:
    km = KMeans(n_clusters=i).fit(df.loc[:, ['lat', 'lon']].values)
    
    res.append(km.inertia_)
plt.plot(n, res)
```

```python
n = 15
km = KMeans(n_clusters=n).fit(df.loc[:, ['lat', 'lon']].values)
df['cluster_id'] = km.predict(df.loc[:, ['lat', 'lon']])
df.head()
```

```python
# loop over all groups and get sensor pairs with 0 great circle distance
sensor_pairs = []

for i in range(n):
    _df = df.loc[df.cluster_id == i, :]
    dist = distance(_df.loc[:, ['lat', 'lon']].values)
    
    # get distance for uniqe sensor pairs
    df_dist = pd.DataFrame(dist)
    # replace columns and index with sensor id for mapping
    df_dist.columns = _df.loc[:, 'sensor_id']
    df_dist.index = pd.Index(_df.loc[:, 'sensor_id'])
    
    # all possible sensor pairs without diagonal e.g. (0,0)
    combs = list(combinations(df_dist.columns, 2))
    
    # create multiindes tuples
    df_dist = df_dist.stack()
    df_dist.index = df_dist.index.set_names(['s1_id', 's2_id'])
    
    # get rows which match combs (= upper triangular matrix without diagonal)
    idx = df_dist.index
    mask = idx.isin(combs)
    df_dist = df_dist[mask].reset_index()

    
    df_dist.columns = ['s1_id', 's2_id', 'dist']

    _d = df_dist.loc[df_dist.dist == 0, ['s1_id', 's2_id']].values.tolist()
    
    sensor_pairs.append(_d)
    
sensor_pairs = [subitem for item in sensor_pairs for subitem in item]
```

```python
G = nx.from_edgelist(sensor_pairs)
hv.Graph.from_networkx(G, nx.layout.fruchterman_reingold_layout).opts(tools=['hover'])
```
